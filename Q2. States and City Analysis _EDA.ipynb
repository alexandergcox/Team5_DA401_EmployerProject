```python
# Get U.S. cities and U.S. states data
us_states_data = gc.get_us_states()
cities_data = gc.get_cities()
```

```python
# convert the cities and US States data into panda dataframes

cities_df = pd.DataFrame(cities_data)
us_states_df = pd.DataFrame(us_states_data)
```

```python
# look at cities data top 5 rows

cities_df.head()
```

```python
# transpose the cities dataset
transposed_cities_df = cities_df.T

# Print the transposed DataFrame
print(transposed_cities_df)
```

```python
# look at tansposed cities data top 5 rows

transposed_cities_df.head()
```

```python
# Filter by 'US' country code
us_cities_df = transposed_cities_df.loc[transposed_cities_df['countrycode'] == 'US']
us_cities_df.head()
```

```python
# rename the us cities columns

us_cities_df.columns = ['geonameid', 'city_name', 'latitude', 'longitude', 'countrycode', 'city_population', 
                        'timezone', 'city_state_code', 'alternate_names' ]

us_cities_df.head()
```

```python
# us cities information
us_cities_df.info()
```

```python
# look to change datatype for latitutde, longitude, city_population to numerical

# Convert 'city_population' column to integer
us_cities_df['city_population'] = us_cities_df['city_population'].astype(int)

# Convert 'latitude' and 'longitude' columns to float
us_cities_df['latitude'] = us_cities_df['latitude'].astype(float)
us_cities_df['longitude'] = us_cities_df['longitude'].astype(float)

# Print the updated DataFrame with data types
print(us_cities_df.dtypes)
```

```python
# us cities information repeated
us_cities_df.info()
us_cities_df.head()
```

```python
# look at top five rows of US states data frame
us_states_df.head()
```

```python
# Transpose the US states dataframe

transposed_US_States_df = us_states_df.T

# Print the transposed DataFrame
print(transposed_US_States_df)
```

```python
# Look at top 5 rows of transposed us states
transposed_US_States_df.head()
```

```python
# rename the us states columns

transposed_US_States_df.columns = ['city_state_code', 'state_name', 'fips', 'geonameid' ]

transposed_US_States_df.head()
```

```python
# Assuming 'city_state_code' is the common column
US_City_States_df = us_cities_df.merge(transposed_US_States_df, on='city_state_code', how='left')

# Print the merged DataFrame
print(US_City_States_df)
```

```python
# look at the new datafame again
US_City_States_df.head()
```

```python
# List of columns to be removed
columns_to_remove = ['geonameid_x', 'countrycode', 'timezone', 'alternate_names', 'fips', 'geonameid_y']

# Remove the specified columns
US_City_States_df = US_City_States_df.drop(columns=columns_to_remove)

# Print the updated DataFrame
print(US_City_States_df)
```

```python
# look at the new datafame again
US_City_States_df.head()
```

```python
# Create a CSV file as output.
US_City_States_df.to_csv('US_City_States_df.csv')
```

# Remove the duplicate cities where name has multiple states
*** After removing cities where no data on latitude et al were included. (Accounted for ~15% of all impressions), there were several cities which which became duplicated due the city name being in multiple states. Through a count if function and True/False logic forumlas in excel, these duplicates could be identified. However due to the size of the CSV, these will need to be cut in python, removing all 'True' in Unique values column.***

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm 
from statsmodels.formula.api import ols
```

```python
# Import the merged file with duplicate informate
City_duplicates = pd.read_csv('Cities_Duplicates_final.csv')
City_duplicates.head()
```

```python
# Assess the data types across each column
City_duplicates.info()
```

```python
# Create Cities_Unique by filtering out rows with 'True' in 'Unique values' column
Cities_Unique = City_duplicates[City_duplicates['Unique values'] != True]
Cities_Unique.head()
```

```python
# Create a CSV file as output.
Cities_Unique.to_csv('Cities_Unique.csv')
# this csv was then uploaded to tableau to develop visualisations to identify the distribution of sessions 
# across states and cities in the USA. Also by merging underbanking population data by state from the
# Federal Deposit Insurance Corporation website, prioritisation by Change Wholelsale of future campaign focus
# could be recommended. This merging was done in excel. A new csv file called Sessions_USA_Population could then be imported
# for work on Tableau visualisation.
```

```python
Cities_Unique = pd.read_csv('Cities_Unique.csv')
Cities_Unique.head()
```

```python
# assess which states have the largest number of sessions over the campaign

State_Sessions = Cities_Unique.groupby(['state_name']).agg({'total_sessions': 'sum'}).reset_index()

# Sort the DataFrame in descending order by 'total_sessions'
State_Sessions = State_Sessions.sort_values(by='total_sessions', ascending=False)

State_Sessions.head(10)
```

```python
# assess which states have the lowest number of sessions over the campaign

State_Sessions = Cities_Unique.groupby(['state_name']).agg({'total_sessions': 'sum'}).reset_index()

# Sort the DataFrame in descending order by 'total_sessions'
State_Sessions = State_Sessions.sort_values(by='total_sessions', ascending=False)

State_Sessions.tail(10)
```

```python
# assess which cities have the largest number of sessions over the campaign

City_Sessions = Cities_Unique.groupby(['city_name']).agg({'total_sessions': 'sum'}).reset_index()

# Sort the DataFrame in descending order by 'total_sessions'
City_Sessions = City_Sessions.sort_values(by='total_sessions', ascending=False)

City_Sessions.head(10)
```
