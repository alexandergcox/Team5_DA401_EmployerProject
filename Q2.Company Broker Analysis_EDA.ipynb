```python
# Group by and aggregate, for the identified columns x columns and metrics

Company_Audience = Company.groupby(['date', 'audience', 'company']).agg({'impressions': 'sum','clicks': 'sum'}).reset_index()

Company_Audience.head()
```

```python
# Convert the 'audience' column to object type
Company_Audience['audience'] = Company_Audience['audience'].astype(str)
Company_Audience.info()
```

```python
# Define a function to map audiences to company_segments based upon whether they were in single audiences or not
def map_audience_to_segment(row):
    if row['audience'] == '1' and row['company'] not in Company_Audience[Company_Audience['audience'] != '1']['company'].values:
        return 'A1 only'
    elif row['audience'] == '2' and row['company'] not in Company_Audience[Company_Audience['audience'] != '2']['company'].values:
        return 'A2 only'
    elif row['audience'] == '3' and row['company'] not in Company_Audience[Company_Audience['audience'] != '3']['company'].values:
        return 'A3 only'
    elif row['audience'] == '4' and row['company'] not in Company_Audience[Company_Audience['audience'] != '4']['company'].values:
        return 'A4 only'
    elif row['audience'] == '5' and row['company'] not in Company_Audience[Company_Audience['audience'] != '5']['company'].values:
        return 'A5 only'
    elif row['audience'] == '6' and row['company'] not in Company_Audience[Company_Audience['audience'] != '6']['company'].values:
        return 'A6 only'
    else:
        return 'Other'

# Apply the function to create the 'audience_segment' column
Company_Audience['audience_segment'] = Company_Audience.apply(map_audience_to_segment, axis=1)

Company_Audience.head()
```

```python
# Replace 'other' with 'Multi-Aud Company' in the 'audience_segment' column
Company_Audience['audience_segment'] = Company_Audience['audience_segment'].replace('Other', 'Multi-Aud Company')

# Check the updated DataFrame
Company_Audience.head()
```
